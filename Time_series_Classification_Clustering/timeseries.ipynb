Ricerca dei segmenti più lunghi senza buchi temporali superiori a 15 minuti

```python
print("\n--- GAP ANALYSIS ---")

# Calcoliamo la differenza di tempo tra una riga e l'altra
traffic_data['delta'] = traffic_data.index.to_series().diff()

# Definiamo un "buco" come qualsiasi salto superiore a 15 minuti
# (Sotto i 15 minuti si scgelie di interpolare, sopra no)
gap_threshold = pd.Timedelta(minutes=15)
is_gap = traffic_data['delta'] > gap_threshold

# Creiamo un ID per ogni gruppo di dati continui
# Ogni volta che c'è un buco, incrementiamo l'ID del gruppo
group_ids = is_gap.cumsum()

# Raggruppiamo per gruppo e calcoliamo inizio, fine e durata
segments = traffic_data.groupby(group_ids).apply(
    lambda x: pd.Series({
        'start': x.index.min(),
        'end': x.index.max(),
        'duration': x.index.max() - x.index.min(),
        'points': len(x)
    })
)

# Ordiniamo per durata decrescente
longest_segments = segments.sort_values('duration', ascending=False).head(5)

print("TOP 5 SEGMENTI CONTINUI (Senza buchi > 15 min):")
print(longest_segments)
```

Reindexing per la gestione dei buchi temporali

```python
traffic_data_full_idx = pd.date_range(start=traffic_data.index.min(), end=traffic_data.index.max(), freq='min')

print(f"Righe X originali: {len(traffic_data)}")
print(f"Righe X target: {len(traffic_data_full_idx)}")

# Reindicizziamo traffic_data
traffic_data = traffic_data.reindex(traffic_data_full_idx)
```

```python
traffic_data['total_flows'] = traffic_data[[
    'npacketsverylow','npacketslow','npacketsmedium','npacketshigh','npacketsveryhigh'
]].sum(axis=1, min_count=1)
```

Nuova serie con interpolazione su buchi <15 min e azzeramento dati nei buchi più grandi 

```python
traffic_1min = traffic_data.copy()

# Interpoliamo i buchi piccoli (< 15 min) per avere continuità nel traffico
traffic_1min['total_flows'] = traffic_1min['total_flows'].interpolate(method='linear', limit=15)
# Riempiamo con 0 i buchi enormi (es. 11 giorni) per dire "traffico zero"
traffic_1min['total_flows'] = traffic_1min['total_flows'].fillna(0)
```

Resample a 60 minuti per limitare gap temporali

```python
import numpy as np

# Aggregazione su 60 minuti (min_count=1 per mantenere NaN nei buchi)
traffic_hours = traffic_data[['total_flows']].resample('60min').sum(min_count=1)

plt.figure(figsize=(15,6))
plt.plot(traffic_hours.index, traffic_hours['total_flows'], color='blue', linewidth=1.5)
plt.title("Numero totale di flussi per ora")
plt.xlabel("Timestamp")
plt.ylabel("Flussi")
plt.show()

# Lavoriamo su una copia
df_fix = traffic_hours.copy()

print("RILEVAMENTO E CORREZIONE DATI PARZIALI")

# Calcolo della mediana storica per ogni (Giorno, Ora) da usare come riferimento
reference_profile = df_fix.groupby([df_fix.index.dayofweek, df_fix.index.hour])['total_flows'].transform('median')

# soglia per identificare i dati parziali (artefatti)
threshold_ratio = 0.50 
is_too_low = df_fix['total_flows'] < (reference_profile * threshold_ratio)

print(f"Punti sospetti trovati (dati parziali): {is_too_low.sum()}")

# Visualizzazione pre-correzione
plt.figure(figsize=(15, 6))
plt.plot(df_fix.index, df_fix['total_flows'], color='gray', alpha=0.5, label='Originale')
plt.scatter(df_fix.index[is_too_low], df_fix.loc[is_too_low, 'total_flows'], color='red', s=30, label='Artefatti Rilevati', zorder=5)
plt.title("Identificazione Ore con Registrazione Parziale")
plt.legend()
plt.show()

# Set a NaN dei valori "sporchi" (artefatti).
df_fix.loc[is_too_low, 'total_flows'] = np.nan

# Calcolo del profilo medio (media per ogni (GiornoSettimana, Ora) escludendo i NaN degli artefatti).
profile_means = df_fix.groupby([df_fix.index.dayofweek, df_fix.index.hour])['total_flows'].transform('mean')

# Riempimento artefatti con il profilo medio
df_fix.loc[is_too_low, 'total_flows'] = df_fix.loc[is_too_low, 'total_flows'].fillna(profile_means)

# Riempimento dei buchi reali piccoli (<= 4 ore) con il profilo medio
# Uso dell'interpolazione lineare solo per creare una maschera che identifichi i buchi piccoli.
# Se interpolate(limit=4) restituisce un numero, il buco è piccolo. Se restituisce NaN, è grande.
mask_small_holes = df_fix['total_flows'].interpolate(method='linear', limit=4).notna() & df_fix['total_flows'].isna()

print(f"Buchi reali piccoli (<=4h) riempiti con media: {mask_small_holes.sum()}")

# Applicazione del profilo medio usando la maschera
df_fix.loc[mask_small_holes, 'total_flows'] = profile_means[mask_small_holes]

# Riempimento Buchi Grandi (> 4 ore) (Zero)
df_fix['total_flows'] = df_fix['total_flows'].fillna(0)

# Confronto finale
plt.figure(figsize=(15, 8))

# Visualizzazione dati vecchi (con i crolli e i buchi a 0 per confronto)
plt.plot(traffic_hours.index, traffic_hours['total_flows'].fillna(0), color='red', alpha=0.3, label='Prima (Con Crolli)')

# Visualizzazione profilo ricostruito
plt.plot(df_fix.index, df_fix['total_flows'], color='blue', linewidth=1.5, label='Dopo (Smart Fix + Zeri)')

# Evidenziazione dei punti riparati
plt.scatter(df_fix.index[is_too_low], df_fix.loc[is_too_low, 'total_flows'], color='green', s=30, label='Punti Ricostruiti (Profilo)', zorder=5)

plt.title("Risultato Finale: Artefatti Corretti vs Buchi a Zero")
plt.ylabel("Flussi")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Sostituzione del dataset originale con quello corretto
traffic_hours = df_fix
```

Decomposizione della serie temporale

```python
from statsmodels.tsa.seasonal import STL

ts = traffic_hours_clean['total_flows']

stl = STL(ts, period=24, robust=True)
result = stl.fit()

# Calcolo soglie anomalie (3 Sigma)
resid = result.resid
sigma = resid.std()
mu = resid.mean()
upper_threshold = mu + 3 * sigma
lower_threshold = mu - 3 * sigma

# Individuazione anomalie
anomalies = ts[ (resid > upper_threshold) | (resid < lower_threshold) ]
print(f"Anomalie rilevate (> 3 sigma): {len(anomalies)}")

# Plotting
fig, axes = plt.subplots(4, 1, figsize=(15, 12), sharex=True)

# Originale
axes[0].plot(ts, label='Traffico Orario', color='gray', alpha=0.7)
axes[0].set_title('1. Serie Originale (Traffico UGR)', fontsize=14)
axes[0].legend(loc='upper left')

# Trend
axes[1].plot(result.trend, label='Trend (Low Frequency)', color='#d35400', linewidth=2)
axes[1].set_title('2. Trend (Assorbe variazioni settimanali)', fontsize=14)
axes[1].legend(loc='upper left')

# Stagionalità
axes[2].plot(result.seasonal, label='Stagionalità (Periodo=24h)', color='#27ae60')
axes[2].set_title('3. Stagionalità Giornaliera', fontsize=14)
axes[2].legend(loc='upper left')

# Residuo
axes[3].scatter(resid.index, resid, label='Residuo', color='#2980b9', s=15, alpha=0.6)
# Linee di soglia rossa
axes[3].axhline(upper_threshold, color='red', linestyle='--', linewidth=1.5, label='Soglia Anomalia (3σ)')
axes[3].axhline(lower_threshold, color='red', linestyle='--', linewidth=1.5)
axes[3].axhline(0, color='black', linewidth=1) # Asse zero

# Evidenziazione anomalie con cerchi rossi
if not anomalies.empty:
    axes[3].scatter(anomalies.index, resid.loc[anomalies.index], color='red', s=50, edgecolors='black', label='Anomalie Rilevate')

axes[3].set_title('4. Residui e Rilevamento Anomalie', fontsize=14)
axes[3].legend(loc='upper left')

plt.tight_layout()
plt.show()
```

Test ADF

```python
from statsmodels.tsa.stattools import adfuller

# ADF test
adf_test = adfuller(train['total_flows'])

# Risultati in formato leggibile
adf_results = pd.Series({
    "ADF Test Statistic": adf_test[0],
    "p-value": adf_test[1],
    "Lags Used": adf_test[2],
    "Number of Observations": adf_test[3],
    "Critical Value (1%)": adf_test[4]["1%"],
    "Critical Value (5%)": adf_test[4]["5%"],
    "Critical Value (10%)": adf_test[4]["10%"]
})

print(adf_results)

# Interpretazione rapida
if adf_test[1] < 0.05:
    print("\n✅ Serie stazionaria: ipotesi nulla rifiutata")
else:
    print("\n⚠️ Serie non stazionaria: considerare differenziazione")
```

ADF per serie differenziata (d=1)

```python
# Preparazione serie (differenza e dropna)
series_diff = train['total_flows'].diff().dropna()

# ADF
adf_test = adfuller(series_diff)

# Calcolo Varianza
variance = np.var(series_diff)

# Risultati in formato leggibile
adf_results = pd.Series({
    "ADF Test Statistic": adf_test[0],
    "p-value": adf_test[1],
    "Lags Used": adf_test[2],
    "Number of Observations": adf_test[3],
    "Critical Value (1%)": adf_test[4]["1%"],
    "Critical Value (5%)": adf_test[4]["5%"],
    "Critical Value (10%)": adf_test[4]["10%"],
    "Variance": variance
})

# Stampa
print("--- RISULTATI TEST STAZIONARIETÀ ---")
print(adf_results)

# Interpretazione Automatica
print("\n--- VERDETTO ---")
if adf_test[1] < 0.05:
    print("✅ Serie Stazionaria (p < 0.05).")
    print(f"La varianza è stabile a: {variance:.2f}")
else:
    print("⚠️ Serie NON Stazionaria.")
```

Scelta del modello

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd

print("\n--- AVVIO CONFRONTO MODELLI CANDIDATI (Fit + Test) ---")

candidates = [
    {'name': '1. Baseline', 'order': (0,1,1), 'seas': (0,1,1,24)},
    {'name': '2. Test AR',   'order': (1,1,1), 'seas': (0,1,1,24)},
    {'name': '3. Test SAR',      'order': (0,1,1), 'seas': (1,1,1,24)},
    {'name': '4. Solo d=1', 'order': (0,1,1), 'seas': (2,0,0,24)},
    {'name': '5. Solo D=1', 'order': (2,0,0), 'seas': (0,1,1,24)}
]

comparison_results = []
airline_fitted_model = None

# Target Test
target_test = test['total_flows'] 

for cand in candidates:
    print(f"Testing: {cand['name']}...", end=" ")
    try:
        # Addestramento
        temp_model = SARIMAX(train['total_flows'], 
                             order=cand['order'], 
                             seasonal_order=cand['seas'], 
                             trend='n', 
                             enforce_stationarity=False, 
                             enforce_invertibility=False)
        temp_res = temp_model.fit(disp=False)
        
        # Previsione e Metriche
        forecast = temp_res.get_forecast(steps=len(target_test))
        pred = forecast.predicted_mean.clip(lower=0)
        
        r2 = r2_score(target_test, pred)
        rmse = np.sqrt(mean_squared_error(target_test, pred))
        mape = np.mean(np.abs((target_test - pred) / target_test)) * 100
        
        comparison_results.append({
            'Modello': cand['name'],
            'AIC': round(temp_res.aic, 2),
            'R2': round(r2, 4),
            'MAPE': f"{mape:.2f}%",
            'RMSE': round(rmse, 0)
        })

        # Salvataggio modello migliore (Baseline)
        if cand['name'] == '1. Baseline':
            airline_fitted_model = temp_res
            print("(Salvato in memoria)", end=" ")
        
        print("Fatto.")
        
    except Exception as e:
        print(f"Errore su {cand['name']}: {e}")

# Stampa Tabella
comp_df = pd.DataFrame(comparison_results)
print("\n--- TABELLA DI CONFRONTO FINALE ---")
print(comp_df)
print("-" * 50)
```

```python
# Configurazione Grafici
plt.style.use('seaborn-v0_8-whitegrid')

print("--- ADDESTRAMENTO E VALIDAZIONE MODELLO SARIMA DEFINITIVO ---")

# Verifica dimensioni
print(f"Training su: {len(train['total_flows'])} ore")
print(f"Testing su:  {len(test['total_flows'])} ore")

# Utilizzo il modello 'Airline' salvato in memoria
if airline_fitted_model is not None:
    print("Utilizzo il modello 'Airline' già addestrato nel confronto.")
    results = airline_fitted_model
else:
    # Se non presente, addestro il modello finale
    print("Modello non in memoria. Riaddestro il modello finale...")
    model = SARIMAX(train['total_flows'], order=(0,1,1), seasonal_order=(0,1,1,24), trend='n')
    results = model.fit(disp=False)

# Configurazione variabili per i log
ORDER = (0, 1, 1)
SEASONAL_ORDER = (0, 1, 1, 24)

# Previsione sul Test Set
forecast_steps = len(test)
forecast_res = results.get_forecast(steps=forecast_steps)

# Estrazione previsioni e intervalli di confidenza
pred = forecast_res.predicted_mean
conf = forecast_res.conf_int()

# Correzione valori negativi
pred = pred.clip(lower=0)
lower_conf = conf.iloc[:, 0].clip(lower=0)
upper_conf = conf.iloc[:, 1]

# Confronto previsioni con realtà
rmse = np.sqrt(mean_squared_error(test['total_flows'], pred))
mae = mean_absolute_error(test['total_flows'], pred)
r2 = r2_score(test['total_flows'], pred)
mape = np.mean(np.abs((test['total_flows'] - pred) / test['total_flows'])) * 100

print(f"\n--- RISULTATI TEST SET (7 GIORNI) ---")
print(f"R2 Score: {r2:.4f} (Obiettivo > 0.7)")
print(f"RMSE:     {rmse:.2f} Flussi")
print(f"MAE:      {mae:.2f} Flussi")
print(f"MAPE:     {mape:.2f}% (Errore percentuale medio)")
print(f"AIC del modello: {results.aic}")

# Visualizzazione risultati
plt.figure(figsize=(12, 8))

# Storico (Zoom ultimi giorni di train)
zoom_days = 5
plt.plot(train.index[-(24*zoom_days):], train.iloc[-(24*zoom_days):], label='Storico (Train)', color='gray', alpha=0.5)

# Realtà Test
plt.plot(test.index, test, label='Realtà (Test Set)', color='black', linewidth=2)

# Previsione
plt.plot(pred.index, pred, label='Previsione SARIMA', color='#27ae60', linewidth=2, linestyle='--')

# Intervallo Confidenza
plt.fill_between(pred.index, lower_conf, upper_conf, color='#27ae60', alpha=0.2, label='Confidenza 95%')

plt.title(f"Forecasting Traffico Reale - R2: {r2:.2f}")
plt.ylabel("Flussi Totali / Ora")
plt.xlabel("Data")
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Analisiamo i residui sul test set
residuals_test = test['total_flows'] - pred
threshold = residuals_test.std() * 3

plt.figure(figsize=(15, 6))
plt.plot(residuals_test.index, residuals_test, color='purple', label='Residuo (Reale - Predetto)')

# Soglie
plt.axhline(threshold, color='red', linestyle='--', label=f'+3 Sigma ({threshold:.0f})')
plt.axhline(-threshold, color='red', linestyle='--', label='-3 Sigma')
plt.axhline(0, color='black', linewidth=1)

# Evidenziamo Anomalie
anomalies = residuals_test[residuals_test > threshold]
if not anomalies.empty:
    plt.scatter(anomalies.index, anomalies, color='red', s=50, zorder=5, label='Anomalie Rilevate')

plt.title("Analisi Residui sul Test Set (Scala Reale)")
plt.ylabel("Errore (Flussi)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Numero di anomalie rilevate (> 3 sigma): {len(anomalies)}")

results.plot_diagnostics(figsize=(15, 12))
plt.show()
```
