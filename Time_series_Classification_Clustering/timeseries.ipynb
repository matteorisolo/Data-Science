Caricamento del dataset e prime verifiche

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "../datasets/UGR16v4.Xtrain.csv"
traffic_data = pd.read_csv(file_path)

traffic_data.info()
print(traffic_data.head())
traffic_data.describe()
```

Elenco delle colonne

```python
list(traffic_data.columns)
```

Verifica conteggio flussi

```python
packet_cols = [
    'npacketsverylow',
    'npacketslow',
    'npacketsmedium',
    'npacketshigh',
    'npacketsveryhigh'
]

byte_cols = [
    'nbytesverylow',
    'nbyteslow',
    'nbytesmedium',
    'nbyteshigh',
    'nbytesveryhigh'
]

traffic_data['total_packets'] = traffic_data[packet_cols].sum(axis=1)
traffic_data['total_bytes'] = traffic_data[byte_cols].sum(axis=1)

traffic_data[['total_packets', 'total_bytes']].head(10)
```

Parsing del timestamp e setting dell'indice

```python
traffic_data['timestamp'] = pd.to_datetime(
    traffic_data['Row'].astype(str),
    format='%Y%m%d%H%M'
)

traffic_data.set_index('timestamp', inplace=True)
```

Ordinamento e verifica

```python
traffic_data.sort_index(inplace=True)
traffic_data.index.min(), traffic_data.index.max()
```

Controllo NaN

```python
traffic_data.isna().sum().sort_values(ascending=False)
```

Controllo duplicati di riga

```python
traffic_data.duplicated().sum()
```

Verifica continuità della serie

```python
# Calcolo la differenza tra un timestamp e quello precedente
diffs = traffic_data.index.to_series().diff()

# Conto quante volte ogni differenza appare
diff_counts = diffs.value_counts().sort_index()

# Mostro le prime 10 differenze più frequenti
with pd.option_context('display.max_rows', None):
    print(diff_counts)
```

Reindexing per la gestione dei buchi temporali

```python
traffic_data_full_idx = pd.date_range(start=traffic_data.index.min(), end=traffic_data.index.max(), freq='min')

print(f"Righe X originali: {len(traffic_data)}")
print(f"Righe X target: {len(traffic_data_full_idx)}")

# Reindicizziamo traffic_data
traffic_data = traffic_data.reindex(traffic_data_full_idx)
```

Interpolazione su buchi <15 min e azzeramento dati nei buchi più grandi 

```python
# Interpoliamo i buchi piccoli (< 15 min) per avere continuità nel traffico
traffic_data.interpolate(method='linear', limit=15, inplace=True)
# Riempiamo con 0 i buchi enormi (es. 11 giorni) per dire "traffico zero"
traffic_data.fillna(0, inplace=True)
```

Verifica assenza dei buchi dopo reindexing, interpolazione e fill

```python
# Calcolo la differenza tra un timestamp e quello precedente
diffs = traffic_data.index.to_series().diff()

# Conto quante volte ogni differenza appare
diff_counts = diffs.value_counts().sort_index()

# Mostro le prime 10 differenze più frequenti
with pd.option_context('display.max_rows', None):
    print(diff_counts)
```

Creazione serie target (Numero di flussi)

```python
# Somma di tutti i flussi per minuto
traffic_data['total_flows'] = traffic_data[['npacketsverylow','npacketslow','npacketsmedium','npacketshigh','npacketsveryhigh']].sum(axis=1)
```

Grafico preliminare della serie

```python
plt.figure(figsize=(15,5))
plt.plot(traffic_data.index, traffic_data['total_flows'], color='blue')
plt.title("Numero totale di flussi per minuto")
plt.xlabel("Timestamp")
plt.ylabel("Flussi")
plt.show()
```
